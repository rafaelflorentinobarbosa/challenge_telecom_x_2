- Transformando as vari√°veis categ√≥ricas em formato num√©rico para torn√°-las compat√≠veis com algoritmos de machine learning

```python
df_tratado['gender'] = df_tratado['gender'].map({'Female': 0, 'Male': 1})

# Identificando as outras colunas categ√≥ricas (que ainda s√£o do tipo 'object')
categorical_cols = df_tratado.select_dtypes(include=['object']).columns

# Aplicando o one-hot encoding nas colunas restantes
df_encoded = pd.get_dummies(df_tratado, columns=categorical_cols, drop_first=True, dtype=int)

# Exibindo as informa√ß√µes do DataFrame ap√≥s todas as transforma√ß√µes
print("\nTipos de dados ap√≥s a codifica√ß√£o:")
print(df_encoded.info())
```

# Calcular Evas√£o

```python
# Calcule a contagem de clientes para cada classe na coluna 'Churn'
churn_counts = df_encoded['Churn'].value_counts()

# Calcule a propor√ß√£o em porcentagem
total_customers = len(df_encoded)
churn_proportion = (churn_counts / total_customers) * 100

# Exiba os resultados
print("Contagem de clientes por classe (0 = Permaneceu, 1 = Evadiu):")
print(churn_counts)

print("\nPropor√ß√£o de clientes por classe:")
print(churn_proportion)

# Avalie o desequil√≠brio das classes
# O desequil√≠brio √© considerado significativo se a propor√ß√£o da classe minorit√°ria for muito pequena
if churn_proportion[1] < 30: # Usamos 30% como um ponto de refer√™ncia comum para desequil√≠brio
    print("\nATEN√á√ÉO: H√° um desequil√≠brio significativo entre as classes.")
    print("Isso pode impactar o desempenho do modelo preditivo.")
    print("Considere t√©cnicas de balanceamento, como oversampling (SMOTE) ou undersampling.")
else:
    print("\nN√£o h√° um desequil√≠brio significativo entre as classes.")
```

# Separa√ß√£o de dados

```python
from sklearn.model_selection import train_test_split

# Defina as vari√°veis preditoras (X) e a vari√°vel alvo (y)
# 'X' s√£o todas as colunas, exceto 'Churn'
# 'y' √© a coluna 'Churn'
X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']

# Divida os dados em conjuntos de treino e teste
# test_size=0.3 aloca 30% dos dados para teste e 70% para treino
# random_state=42 garante que a divis√£o seja a mesma em todas as execu√ß√µes
# stratify=y garante que a propor√ß√£o das classes de 'Churn' seja mantida em ambos os conjuntos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Exiba o tamanho dos conjuntos para confirmar a divis√£o
print("Tamanho do conjunto X_train:", X_train.shape)
print("Tamanho do conjunto X_test:", X_test.shape)
print("Tamanho do conjunto y_train:", y_train.shape)
print("Tamanho do conjunto y_test:", y_test.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler

# Defina as vari√°veis preditoras (X) e a vari√°vel alvo (y)
X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']

# Divida os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)


# --- Modelo 1: Regress√£o Log√≠stica (requer normaliza√ß√£o) ---
# Justificativa: A Regress√£o Log√≠stica √© um modelo linear baseado em coeficientes,
# e seu desempenho √© afetado pela escala das vari√°veis. Para evitar que
# vari√°veis com grandes faixas de valores dominem o modelo, √© essencial
# padronizar os dados.

print("Treinando o modelo de Regress√£o Log√≠stica...")

# Identifique as colunas num√©ricas que precisam de padroniza√ß√£o
numerical_cols = ['tenure', 'Charges.Monthly', 'Charges.Total', 'Contas_Diarias']

# Crie uma c√≥pia dos conjuntos de treino e teste para a normaliza√ß√£o
X_train_lr = X_train.copy()
X_test_lr = X_test.copy()

# Crie uma inst√¢ncia do StandardScaler
scaler = StandardScaler()

# Ajuste o scaler aos dados de treino e transforme os dados de treino e teste
X_train_lr[numerical_cols] = scaler.fit_transform(X_train_lr[numerical_cols])
X_test_lr[numerical_cols] = scaler.transform(X_test_lr[numerical_cols])

# Crie e treine o modelo de Regress√£o Log√≠stica
model_lr = LogisticRegression(random_state=42)
model_lr.fit(X_train_lr, y_train)
print("Modelo de Regress√£o Log√≠stica treinado com sucesso!")
print("-" * 50)


# --- Modelo 2: Random Forest (n√£o requer normaliza√ß√£o) ---
# Justificativa: O Random Forest √© um modelo baseado em √°rvores de decis√£o.
# As √°rvores de decis√£o funcionam com base em "limiares" (thresholds) de vari√°veis,
# e n√£o na dist√¢ncia entre os pontos. Portanto, elas n√£o s√£o sens√≠veis √† escala
# dos dados, e a normaliza√ß√£o n√£o √© necess√°ria.

print("Treinando o modelo de Random Forest...")

# Crie e treine o modelo Random Forest
model_rf = RandomForestClassifier(random_state=42)
model_rf.fit(X_train, y_train)
print("Modelo de Random Forest treinado com sucesso!")
```

```python
# Importe as m√©tricas de avalia√ß√£o
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler

# O c√≥digo a seguir assume que a vari√°vel df_encoded j√° est√° carregada e pronta para uso.
# Defina as vari√°veis preditoras (X) e a vari√°vel alvo (y)
X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']

# Divida os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)


# --- Treinamento do Modelo de Regress√£o Log√≠stica ---
# A Regress√£o Log√≠stica requer normaliza√ß√£o dos dados num√©ricos.
# Identifique as colunas num√©ricas (exceto as bin√°rias criadas pelo one-hot encoding)
numerical_cols = ['tenure', 'Charges.Monthly', 'Charges.Total', 'Contas_Diarias']
# Crie uma c√≥pia dos conjuntos de treino e teste para a normaliza√ß√£o
X_train_lr = X_train.copy()
X_test_lr = X_test.copy()
# Crie uma inst√¢ncia do StandardScaler
scaler = StandardScaler()
# Ajuste o scaler aos dados de treino e transforme os dados de treino e teste
X_train_lr[numerical_cols] = scaler.fit_transform(X_train_lr[numerical_cols])
X_test_lr[numerical_cols] = scaler.transform(X_test_lr[numerical_cols])
# Crie e treine o modelo de Regress√£o Log√≠stica
model_lr = LogisticRegression(random_state=42)
model_lr.fit(X_train_lr, y_train)


# --- Treinamento do Modelo de Random Forest ---
# O Random Forest n√£o requer normaliza√ß√£o.
# Crie e treine o modelo Random Forest
model_rf = RandomForestClassifier(random_state=42)
model_rf.fit(X_train, y_train)


# --- Avalia√ß√£o e Visualiza√ß√£o da Matriz de Confus√£o para a Regress√£o Log√≠stica ---
print("=== Avalia√ß√£o do Modelo de Regress√£o Log√≠stica ===")
y_pred_lr = model_lr.predict(X_test_lr)
print(f"Acur√°cia: {accuracy_score(y_test, y_pred_lr):.4f}")
print(f"Precis√£o: {precision_score(y_test, y_pred_lr):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_lr):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred_lr):.4f}")

# Crie e visualize a matriz de confus√£o
cm_lr = confusion_matrix(y_test, y_pred_lr)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confus√£o - Regress√£o Log√≠stica')
plt.xlabel('R√≥tulo Previsto')
plt.ylabel('R√≥tulo Verdadeiro')
plt.show()


# --- Avalia√ß√£o e Visualiza√ß√£o da Matriz de Confus√£o para o Random Forest ---
print("\n=== Avalia√ß√£o do Modelo de Random Forest ===")
y_pred_rf = model_rf.predict(X_test)
print(f"Acur√°cia: {accuracy_score(y_test, y_pred_rf):.4f}")
print(f"Precis√£o: {precision_score(y_test, y_pred_rf):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_rf):.4f}")
print(f"F1-Score: {f1_score(y_test, y_pred_rf):.4f}")

# Crie e visualize a matriz de confus√£o
cm_rf = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confus√£o - Random Forest')
plt.xlabel('R√≥tulo Previsto')
plt.ylabel('R√≥tulo Verdadeiro')
plt.show()
```

# An√°lise de Import√¢ncia das Vari√°veis

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler

# Defina as vari√°veis preditoras (X) e a vari√°vel alvo (y)
X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']

# Divida os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)


# --- Treinamento do Modelo de Regress√£o Log√≠stica (com normaliza√ß√£o) ---
numerical_cols = ['tenure', 'Charges.Monthly', 'Charges.Total', 'Contas_Diarias']
X_train_lr = X_train.copy()
X_test_lr = X_test.copy()
scaler = StandardScaler()
X_train_lr[numerical_cols] = scaler.fit_transform(X_train_lr[numerical_cols])
X_test_lr[numerical_cols] = scaler.transform(X_test_lr[numerical_cols])
model_lr = LogisticRegression(random_state=42)
model_lr.fit(X_train_lr, y_train)


# --- Treinamento do Modelo de Random Forest (sem normaliza√ß√£o) ---
model_rf = RandomForestClassifier(random_state=42)
model_rf.fit(X_train, y_train)


# --- An√°lise de Import√¢ncia das Vari√°veis para a Regress√£o Log√≠stica ---
print("=== Import√¢ncia das Vari√°veis (Regress√£o Log√≠stica) ===")
# Obtenha os coeficientes do modelo
feature_importance_lr = pd.DataFrame({
    'Vari√°vel': X_train_lr.columns,
    'Coeficiente': model_lr.coef_[0]
})
# Ordene por valor absoluto do coeficiente para ver as mais impactantes
feature_importance_lr['Abs_Coeficiente'] = np.abs(feature_importance_lr['Coeficiente'])
feature_importance_lr = feature_importance_lr.sort_values(by='Abs_Coeficiente', ascending=False)
print(feature_importance_lr[['Vari√°vel', 'Coeficiente']])
print("-" * 50)


# --- An√°lise de Import√¢ncia das Vari√°veis para o Random Forest ---
print("\n=== Import√¢ncia das Vari√°veis (Random Forest) ===")
# Obtenha a import√¢ncia das vari√°veis do modelo Random Forest
feature_importance_rf = pd.DataFrame({
    'Vari√°vel': X_train.columns,
    'Import√¢ncia': model_rf.feature_importances_
})
# Ordene por import√¢ncia em ordem decrescente
feature_importance_rf = feature_importance_rf.sort_values(by='Import√¢ncia', ascending=False)
print(feature_importance_rf)
```

Prepara√ß√£o dos Dados
Antes da modelagem, realizamos as seguintes etapas para preparar os dados:

Divis√£o em Treino e Teste: Garantimos que os modelos fossem avaliados em dados n√£o vistos, dividindo o conjunto de dados em 70% para treino e 30% para teste. A divis√£o foi feita de forma estratificada para manter a propor√ß√£o das classes de evas√£o em ambos os conjuntos.

Codifica√ß√£o das Vari√°veis Categ√≥ricas: Transformamos as vari√°veis de texto em formato num√©rico, utilizando o one-hot encoding para a maioria das colunas e uma codifica√ß√£o manual para a vari√°vel gender.

Normaliza√ß√£o com StandardScaler: Essa etapa foi aplicada especificamente para o modelo de Regress√£o Log√≠stica, que √© sens√≠vel √† escala dos dados. A normaliza√ß√£o transforma as vari√°veis num√©ricas para que tenham m√©dia 0 e desvio padr√£o 1, garantindo que n√£o haja vari√°veis com escalas dominantes.

Modelos Utilizados
Foram escolhidos dois modelos principais para este desafio:

1. Regress√£o Log√≠stica

Motivo da escolha: √â um modelo linear, simples e interpret√°vel, que serve como um bom ponto de partida. Ele nos permite entender a contribui√ß√£o de cada vari√°vel (atrav√©s de seus coeficientes) para a previs√£o.

Normaliza√ß√£o: Essencial. As vari√°veis num√©ricas foram normalizadas para garantir que o modelo n√£o fosse enviesado por diferen√ßas de escala.

Resultados: [Preencha com a sua an√°lise cr√≠tica baseada nas m√©tricas e matriz de confus√£o.]

2. Random Forest

Motivo da escolha: √â um modelo de ensemble robusto, que geralmente oferece melhor capacidade de generaliza√ß√£o e reduz o risco de overfitting em compara√ß√£o com modelos mais simples como a √°rvore de decis√£o.

Normaliza√ß√£o: N√£o necess√°ria. Modelos baseados em √°rvores n√£o s√£o sens√≠veis √† escala dos dados.

Resultados: [Preencha com a sua an√°lise cr√≠tica baseada nas m√©tricas e matriz de confus√£o.]

Visualiza√ß√µes e M√©tricas
